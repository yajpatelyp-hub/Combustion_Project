```python
import numpy as np
import matplotlib.pyplot as plt
import cantera as ct
from Project_3_2 import solve_pfr_cantera

# --------------------------------------------------------------
# 1. Problem setup for verification
# --------------------------------------------------------------
mech = "gri30.yaml"
X0 = "N2:1"          # inert gas
T0 = 300.0           # inlet temperature
Tw = 1200.0          # constant wall temperature
p0 = ct.one_atm
mdot = 0.1           # kg/s
D = 0.05             # m
x_end = 1.0          # reactor length in meters

# --------------------------------------------------------------
# 2. Define constant h and constant cp for analytic test
# --------------------------------------------------------------
h_const_value = 50.0       # W/m^2-K (chosen constant)
cp_const_value = 1000.0    # J/kg-K (chosen constant)

def h_const(T, Y, p, rho, u):
    return h_const_value

def cp_const(T):
    return cp_const_value

# --------------------------------------------------------------
# 3. Run the solver with overrides
# --------------------------------------------------------------
sol = solve_pfr_cantera(
    mechfile = mech,
    X0 = X0,
    T0 = T0,
    p0 = p0,
    mdot = mdot,
    D = D,
    Tw = Tw, 
    x_end = x_end,
    h_override = h_const,     # constant h
    cp_override = cp_const,   # constant cp
)

x = sol["x"]
T_num = sol["T"]

# --------------------------------------------------------------
# 4. Compute analytic solution
# --------------------------------------------------------------
Per = np.pi * D
K = (h_const_value * Per) / (mdot * cp_const_value)   # decay constant

T_analytic = Tw + (T0 - Tw) * np.exp(-K * x)

# --------------------------------------------------------------
# 5. Compute error metrics
# --------------------------------------------------------------
abs_err = np.abs(T_num - T_analytic)
max_err = np.max(abs_err)
rmse = np.sqrt(np.mean(abs_err**2))

print("Verification 4.1 — Constant h Analytic Test")
print("------------------------------------------------")
print(f"Constant h used: {h_const_value:.2f} W/m²-K")
print(f"Constant cp used: {cp_const_value:.1f} J/kg-K")
print(f"Decay constant K: {K:.4f} 1/m")
print(f"Max absolute error : {max_err:.6f} K")
print(f"RMSE               : {rmse:.6f} K")
print("------------------------------------------------")

# --------------------------------------------------------------
# 6. Plot numerical vs analytic solution
# --------------------------------------------------------------
plt.figure(figsize=(8,6))
plt.plot(x, T_num, 'b-', linewidth=2, label="Numerical (PFR Solver)")
plt.plot(x, T_analytic, 'r--', linewidth=2.5, label="Analytic Solution")

plt.xlabel("x (m)")
plt.ylabel("Temperature (K)")
plt.title("Verification 4.1 — Constant-h Analytic Check")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cantera as ct
from Project_3_2 import solve_pfr_cantera

# Strongly curved exponential test
h_const_value = 500.0        # large h
cp_const_value = 500.0       # smaller cp
mdot_test = 0.02             # small mdot
D_test = 0.05
Tw = 1200.0
T0 = 300.0
p0 = ct.one_atm
x_end = 1.0

def h_const(T, Y, p, rho, u):
    return h_const_value

def cp_const(T):
    return cp_const_value

sol = solve_pfr_cantera(
    mechfile="gri30.yaml",
    X0="N2:1",
    T0=T0,
    p0=p0,
    mdot=mdot_test,
    D=D_test,
    Tw=Tw,
    x_end=x_end,
    h_override=h_const,
    cp_override=cp_const
)

x = sol["x"]
T_num = sol["T"]

# Compute analytic solution
Per = np.pi * D_test
K = (h_const_value * Per) / (mdot_test * cp_const_value)
T_analytic = Tw + (T0 - Tw) * np.exp(-K * x)

print("Decay constant K:", K)

plt.figure(figsize=(8,6))
plt.plot(x, T_num, 'b-', linewidth=2, label="Numerical")
plt.plot(x, T_analytic, 'r--', linewidth=2, label="Analytic")
plt.grid(True)
plt.xlabel("x (m)")
plt.ylabel("T (K)")
plt.legend()
plt.show()
```
